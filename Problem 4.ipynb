```python
#This file contains the first 25 Fibonacci numbers in its only column. 
#Add the next 25 fibonacci numbers into the DataFrame. 
#A fibonacci number is a sequence where the current number is the sum of previous two numbers in the sequence.
import pandas as pd
fib = pd.read_csv('fibonacci.csv')
fib
```

```python
fib1=pd.Series(0*25,index=range(25,51))

fib1[25]=fib.iloc[23]['Fibonacci']+fib.iloc[24]['Fibonacci']
fib1[26]=fib.iloc[24]['Fibonacci']+fib1[25]
for i in range(27,51):
    fib1[i]=fib1[i-2]+fib1[i-1]
fib1
```

```python
df = pd.concat([fib, fib1])
for i in range(25,51):
    df.iloc[i]['Fibonacci']=df.iloc[i][0]
```

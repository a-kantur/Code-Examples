```python
import pandas as pd
```

```python
movies=pd.read_csv('movies.csv')
```

```python
import random
index = [i for i in range(movies.shape[0])]
random.shuffle(index)
movies.set_index([index]).sort_index()
```

```python
import random as rnd
movies['random sort']=movies.apply(lambda x: rnd.random(), axis=1)
```

```python
movies.sort_values(by='random sort')
```
